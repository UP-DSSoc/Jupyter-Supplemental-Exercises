# I. Logic and Control Flow

#### `if` / `elif` and `else` Statement and Conditionals

- `<` less than
- `>` greater than
- `<=` less than or equal to
- `>=` greater than or equal to
- `==` equal to

```python
a = 0

if a < 0 :
    print("a is a negative number")
elif a > 0 :
    print("a is a positve number")
else:
    print("a is zero")
```

#### `and ` and `or`  Operators

```python
b = 2

if b < 0 or b > 0 :
    print("b is not zero")
else : 
    print ("b is zero")
```

```python
c = 1
d = c%3
e = c%2

if d > 0 and e > 0 :
    print("c is not divisible by 3 and 2")
elif d > 0 or e > 0 :
    print("c is divisible by 3 or 2")
elif d == 0 and e == 0 :
    print("c is divisible by 3 and 2")    
```
